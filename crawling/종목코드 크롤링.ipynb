```python
import FinanceDataReader as fdr
import pandas as pd
from tqdm.notebook import tqdm
import time

from selenium import webdriver
from bs4 import BeautifulSoup
import requests
```

```python
#주식정보 가져오기
df = fdr.StockListing('KRX')

#종목명, 종목코드 데이터프레임 생성

code = df[['Symbol', 'Name']]
```

```python
#종목이름 크롤링

url = 'https://finance.daum.net/domestic/volume?market=KOSPI'

driver = webdriver.Chrome(r'C:\Users\tjddy\ipynb_files\ipynb_files\chromedriver\chromedriver.exe')
driver.get(url)

names = []
```

```python
for page in tqdm(range(1,40)):

    soup = BeautifulSoup(driver.page_source)
    body = soup.find_all('tbody')[1]
    stop = 0
    
    names_soup = body.find_all('a', class_='txt') #a테그에 적힌 종목명 가져오기
    
    
    for name in names_soup: #names_Soup에서 text값만 뽑아서 names에 저장
        stock = name.get_text()
        if 'KODEX' not in stock and '인버스' not in stock and '선물' not in stock: #KODEX, 인버스, 선물 종목 제거
            names.append(stock)
        
        num = len(names)
        
        if num >= 600:
            stop = 1
            break
            
    if stop == 1: #종료조건, 600개 쌓이면
        driver.quit()
        break
        
    #페이지 넘김    
    if page <= 10:
        driver.find_element_by_xpath('//*[@id="boxTradeVolume"]/div[2]/div[1]/div/a' + '[' + str(page) + ']').click()
    else:
        if page % 10 ==0:
            next = 12
        else:
            next = (page%10)+2
        
        driver.find_element_by_xpath('//*[@id="boxTradeVolume"]/div[2]/div[1]/div/a' + '[' + str(next) + ']').click()

    time.sleep(4)

driver.quit()
```

```python
name1 = names[:200]    #상휘
name2 = names[200:400] #성용
name3 = names[400:600] #해준
```

```python
#딕셔너리형태로 변환

codes1 = {}
codes2 = {}
codes3 = {}

for i in name1:
    codes1[i] = code[code['Name']==i][['Symbol']].iloc[0][0]

for i in name2:
    codes2[i] = code[code['Name']==i][['Symbol']].iloc[0][0]
    
for i in name3:
    codes3[i] = code[code['Name']==i][['Symbol']].iloc[0][0]
```

```python
#개수확인

print(len(codes1), len(codes2), len(codes3))
```
